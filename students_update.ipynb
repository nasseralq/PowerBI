```python
def first(First_period_grade):
    if First_period_grade >= 18:
        return "A+"
    elif First_period_grade >=16:
        return "A"
    elif First_period_grade>= 14:
        return "B"
    elif First_period_grade >= 10:
        return "C"
    else:
        return "F"
```

```python
df['First_period_grade']= df.First_period_grade.apply(first)
df
```

```python
df['Second_period_grade']= df.Second_period_grade.apply(first)
df
```

```python
df['Final_grade']= df.Final_grade.apply(first)
df
```
